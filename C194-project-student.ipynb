# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

health = pd.read_csv('framingham.csv')
health
```

```python
nonsmoker=health.loc[health['currentSmoker'] == 0]
nonsmoker
```

```python
groupby_nonsmoker=nonsmoker.groupby('age')['heartRate'].mean()
groupby_nonsmoker
```

```python
smoker=health.loc[health['currentSmoker'] == 0]
smoker
```

```python
groupby_smoker=nonsmoker.groupby('age')['heartRate'].mean()
groupby_smoker
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
plt.figure(figsize=(12,8))

plt.plot(groupby_nonsmoker['age'],groupby_nonsmoker['heartRate'], label='Heart Rate of Nonsmokers',linewidth=3)
plt.plot(groupby_smoker['age'],groupby_smoker['heartRate'], label='Heart Rate of Smokers',linewidth=3)

plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.title('Heart Rate of Smokers and Nonsmokers')
plt.legend()
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
health2 = pd.read_csv('framingham.csv')
coronary_heart_disease=health2.loc[health2['TenYearCHD'] == 1]
coronary_heart_disease
```

```python
group_age=coronary_heart_disease.groupby('age')['TenYearCHD'].count().reset_index()
group_age
```

```python
plt.figure(figsize=(12,8))

plt.scatter(group_age['age'],group_age['TenYearCHD'], label='People with 10 years of CHD',color='red',alpha=0.5,
            s=group_age['TenYearCHD']*50)

plt.xlabel('Age')
plt.ylabel('People with 10 years of CHD')
plt.title('Different ages with 10 years of CHD')
plt.legend()
plt.show()
```
